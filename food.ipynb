```python
#fill all the missing values in numerical columns with their median
numerical_cols=food.select_dtypes(include=['number']).columns
food[numerical_cols]=food[numerical_cols].fillna (food[numerical_cols].median())
```

```python
#fill all the missinig values in categorical columns with the most common value (mode)
categorical_cols=food.select_dtypes(include=['number']).columns
food[categorical_cols]=food[categorical_cols].fillna (food[categorical_cols].mode().iloc[0])
```

```python
#to remove outliers using IQR method for all numerical columns
def remove_outliers_iqr(df):
    numeric_cols=df.select_dtype(include=['float64','int64']).columns
    for col in numeric_cols:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR= (Q3-Q1)
        lower_bound=Q1-1.5*IQR
        upper_bound=Q3+1.5*IQR
        #keeping only new
        df=df[(df[col]>=lower_bound)] & (df[col]<=upper_bound)
        return df #return the modified dataframe
    # applying the function above to my dtaframe
    food_cleaned=remove_outliers_iqr(food)
    #to check results
    print('shape after removing outliers',food_cleaned.shape)
                
```

ENCODING CATEGORICAL DATA INTO NUMERICAL FORM

```python
from sklearn.preprocessing import LabelEncoder
#initialize labelEncoder
Label_Encoder= LabelEncoder()
#select categorical columns
categorical_cols=food.select_dtypes(include=['object']).columns
#Apply label encoding to each categorical columns
for col in categorical_cols:
    food[col]=Label_Encoder.fit_transform(food[col])
    
    food.head()
```

NORMALIZATION  (MIN-MAX SCALING)

```python
from sklearn.preprocessing import MinMaxScaler
#select numerical columns
numerical_cols=food.select_dtypes(include=['float64', 'int64']).columns
#Apply normalization
scaler=MinMaxScaler()
food[numerical_cols]=scaler.fit_transform(food[numerical_cols])
```
